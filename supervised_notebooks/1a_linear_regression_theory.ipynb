## 1. Mathematical Foundations

### The Linear Model

Linear regression models the relationship between:
- **Input features** $X = (x_1, x_2, ..., x_n)$
- **Output target** $y$

using a **linear function**:

$$
\hat{y} = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + ... + \theta_n x_n
$$

Where:
- $\hat{y}$ = predicted value
- $\theta_0$ = intercept (bias term)
- $\theta_1, ..., \theta_n$ = weights (coefficients)
- $x_1, ..., x_n$ = feature values

**Vectorized form:**

$$
\hat{y} = \theta^T X
$$

Where:
- $\theta = [\theta_0, \theta_1, ..., \theta_n]^T$ (parameters)
- $X = [1, x_1, x_2, ..., x_n]^T$ (features with bias term)

**Simple Linear Regression** (one feature):

$$
\hat{y} = \theta_0 + \theta_1 x
$$

This is the equation of a straight line:
- $\theta_0$ = y-intercept (where the line crosses the y-axis)
- $\theta_1$ = slope (how steep the line is)

---

### Cost Function (Mean Squared Error)

**Goal**: Find parameters $\theta$ that make predictions $\hat{y}$ as close as possible to actual values $y$.

**How do we measure "closeness"?** Using the **Mean Squared Error (MSE)**:

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})^2
$$

$$
J(\theta) = \frac{1}{2m} \sum_{i=1}^{m} (\theta^T X^{(i)} - y^{(i)})^2
$$

Where:
- $m$ = number of training examples
- $\hat{y}^{(i)}$ = prediction for example $i$
- $y^{(i)}$ = actual value for example $i$
- $\frac{1}{2}$ = mathematical convenience (cancels out during differentiation)

**Why squared error?**
1. ✅ Penalizes large errors more than small ones
2. ✅ Always positive (no cancellation)
3. ✅ Differentiable (smooth for optimization)
4. ✅ Convex function (single global minimum)

**Matrix form:**

$$
J(\theta) = \frac{1}{2m} (X\theta - y)^T (X\theta - y)
$$

---

### Gradient Descent

**Goal**: Minimize $J(\theta)$ by iteratively adjusting $\theta$.

**Algorithm**:

1. Initialize $\theta$ randomly (or to zeros)
2. Repeat until convergence:
   $$
   \theta_j := \theta_j - \alpha \frac{\partial J(\theta)}{\partial \theta_j}
   $$

Where:
- $\alpha$ = learning rate (step size)
- $\frac{\partial J(\theta)}{\partial \theta_j}$ = partial derivative (gradient)

**Derive the gradient:**

$$
\frac{\partial J(\theta)}{\partial \theta_j} = \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) \cdot x_j^{(i)}
$$

**Vectorized update rule:**

$$
\theta := \theta - \alpha \frac{1}{m} X^T (X\theta - y)
$$

**Intuition:**
- Gradient points in direction of **steepest ascent**
- We subtract it to go **downhill** (minimize cost)
- Learning rate $\alpha$ controls how big each step is

**Gradient Descent Variants:**
1. **Batch Gradient Descent**: Use all training examples (what we'll implement)
2. **Stochastic Gradient Descent (SGD)**: Use one example at a time
3. **Mini-batch Gradient Descent**: Use small batches of examples

---

### Normal Equation (Closed-Form Solution)

**Alternative to gradient descent**: Solve for $\theta$ analytically!

**How?** Take the derivative of $J(\theta)$, set it to zero, and solve:

$$
\frac{\partial J(\theta)}{\partial \theta} = 0
$$

**Result:**

$$
\theta = (X^T X)^{-1} X^T y
$$

This gives the **optimal parameters in one step** (no iteration needed)!

**Proof sketch:**

1. Cost function: $J(\theta) = \frac{1}{2m} (X\theta - y)^T (X\theta - y)$
2. Expand: $J(\theta) = \frac{1}{2m} (\theta^T X^T X \theta - 2\theta^T X^T y + y^T y)$
3. Take derivative: $\frac{\partial J}{\partial \theta} = \frac{1}{m}(X^T X \theta - X^T y)$
4. Set to zero: $X^T X \theta - X^T y = 0$
5. Solve: $\theta = (X^T X)^{-1} X^T y$

**Gradient Descent vs. Normal Equation:**

| Aspect | Gradient Descent | Normal Equation |
|--------|-----------------|----------------|
| Speed | Slow for small datasets | Fast for small datasets |
| Large datasets (n) | Works well | Slow ($O(n^3)$ for matrix inversion) |
| Learning rate | Need to choose $\alpha$ | No hyperparameters |
| Iterations | Many | One step |
| When to use | $n > 10,000$ features | $n < 10,000$ features |

### Simple Linear Regression (One Feature)

First, let's create a simple dataset and visualize it.

```python
# Generate synthetic data: y = 3 + 2*x + noise
np.random.seed(42)
m = 100  # number of examples
X_simple = 2 * np.random.rand(m, 1)  # features in [0, 2]
y_simple = 3 + 2 * X_simple + np.random.randn(m, 1)  # true relationship + noise

# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X_simple, y_simple, alpha=0.6, s=50)
plt.xlabel('X (Feature)', fontsize=12)
plt.ylabel('y (Target)', fontsize=12)
plt.title('Simple Linear Regression Dataset\n(True relationship: y = 3 + 2x)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.show()

print(f"Dataset: {m} examples")
print(f"Feature range: [{X_simple.min():.2f}, {X_simple.max():.2f}]")
print(f"Target range: [{y_simple.min():.2f}, {y_simple.max():.2f}]")
```

## 3. Gradient Descent Implementation

Now let's implement gradient descent from scratch!

```python
class LinearRegressionGD:
    """
    Linear Regression using Gradient Descent (from scratch with NumPy)
    
    Parameters:
    -----------
    learning_rate : float
        Step size for gradient descent (alpha)
    n_iterations : int
        Number of iterations to run gradient descent
    """
    
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.theta = None
        self.cost_history = []
        
    def _add_bias(self, X):
        """Add bias term (column of ones) to X"""
        return np.c_[np.ones((X.shape[0], 1)), X]
    
    def _compute_cost(self, X, y, theta):
        """Compute Mean Squared Error cost function"""
        m = len(y)
        predictions = X @ theta
        cost = (1 / (2 * m)) * np.sum((predictions - y) ** 2)
        return cost
    
    def fit(self, X, y):
        """
        Train the model using gradient descent
        
        Parameters:
        -----------
        X : array-like, shape (m, n)
            Training features
        y : array-like, shape (m, 1)
            Target values
        """
        # Add bias term
        X_b = self._add_bias(X)
        m, n = X_b.shape
        
        # Initialize parameters
        self.theta = np.random.randn(n, 1)
        
        # Gradient descent
        for iteration in range(self.n_iterations):
            # Compute predictions
            predictions = X_b @ self.theta
            
            # Compute gradients
            gradients = (1 / m) * X_b.T @ (predictions - y)
            
            # Update parameters
            self.theta = self.theta - self.learning_rate * gradients
            
            # Track cost
            cost = self._compute_cost(X_b, y, self.theta)
            self.cost_history.append(cost)
            
            # Print progress every 100 iterations
            if iteration % 100 == 0:
                print(f"Iteration {iteration}: Cost = {cost:.4f}")
        
        return self
    
    def predict(self, X):
        """
        Make predictions on new data
        
        Parameters:
        -----------
        X : array-like, shape (m, n)
            Features to predict on
            
        Returns:
        --------
        predictions : array-like, shape (m, 1)
            Predicted values
        """
        X_b = self._add_bias(X)
        return X_b @ self.theta

print("✓ LinearRegressionGD class implemented")
```

### Train the model

```python
# Create and train model
model_gd = LinearRegressionGD(learning_rate=0.1, n_iterations=1000)
model_gd.fit(X_simple, y_simple)

print("\n" + "="*50)
print("GRADIENT DESCENT RESULTS")
print("="*50)
print(f"Learned parameters:")
print(f"  θ₀ (intercept) = {model_gd.theta[0][0]:.4f}")
print(f"  θ₁ (slope)     = {model_gd.theta[1][0]:.4f}")
print(f"\nTrue parameters:")
print(f"  θ₀ (intercept) = 3.0000")
print(f"  θ₁ (slope)     = 2.0000")
print(f"\nFinal cost: {model_gd.cost_history[-1]:.4f}")
```

### Visualize the fit

```python
# Make predictions
X_plot = np.linspace(0, 2, 100).reshape(-1, 1)
y_pred = model_gd.predict(X_plot)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(X_simple, y_simple, alpha=0.6, s=50, label='Training data')
plt.plot(X_plot, y_pred, 'r-', linewidth=2, label=f'Learned line (y = {model_gd.theta[0][0]:.2f} + {model_gd.theta[1][0]:.2f}x)')
plt.plot(X_plot, 3 + 2*X_plot, 'g--', linewidth=2, alpha=0.7, label='True line (y = 3 + 2x)')
plt.xlabel('X (Feature)', fontsize=12)
plt.ylabel('y (Target)', fontsize=12)
plt.title('Linear Regression: Gradient Descent Fit', fontsize=14)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()
```

## 4. Normal Equation Implementation

Now let's implement the closed-form solution!

```python
class LinearRegressionNormal:
    """
    Linear Regression using Normal Equation (closed-form solution)
    """
    
    def __init__(self):
        self.theta = None
        
    def _add_bias(self, X):
        """Add bias term (column of ones) to X"""
        return np.c_[np.ones((X.shape[0], 1)), X]
    
    def fit(self, X, y):
        """
        Train the model using normal equation
        
        θ = (X^T X)^(-1) X^T y
        
        Parameters:
        -----------
        X : array-like, shape (m, n)
            Training features
        y : array-like, shape (m, 1)
            Target values
        """
        X_b = self._add_bias(X)
        
        # Normal equation: θ = (X^T X)^(-1) X^T y
        self.theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
        
        return self
    
    def predict(self, X):
        """
        Make predictions on new data
        """
        X_b = self._add_bias(X)
        return X_b @ self.theta

print("✓ LinearRegressionNormal class implemented")
```

```python
# Train using normal equation
model_normal = LinearRegressionNormal()
model_normal.fit(X_simple, y_simple)

print("\n" + "="*50)
print("NORMAL EQUATION RESULTS")
print("="*50)
print(f"Learned parameters:")
print(f"  θ₀ (intercept) = {model_normal.theta[0][0]:.4f}")
print(f"  θ₁ (slope)     = {model_normal.theta[1][0]:.4f}")
print(f"\nTrue parameters:")
print(f"  θ₀ (intercept) = 3.0000")
print(f"  θ₁ (slope)     = 2.0000")

# Compare with gradient descent
print(f"\n" + "="*50)
print("COMPARISON")
print("="*50)
print(f"Parameter difference:")
print(f"  Δθ₀ = {abs(model_gd.theta[0][0] - model_normal.theta[0][0]):.6f}")
print(f"  Δθ₁ = {abs(model_gd.theta[1][0] - model_normal.theta[1][0]):.6f}")
print(f"\nBoth methods converge to the same solution! ✓")
```

### Visualize cost function landscape (for simple case)

```python
# Create a grid of theta values
theta0_vals = np.linspace(1, 5, 100)
theta1_vals = np.linspace(0, 4, 100)
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

X_b = np.c_[np.ones((X_simple.shape[0], 1)), X_simple]

# Compute cost for each combination
for i, theta0 in enumerate(theta0_vals):
    for j, theta1 in enumerate(theta1_vals):
        theta = np.array([[theta0], [theta1]])
        predictions = X_b @ theta
        cost = (1 / (2 * len(y_simple))) * np.sum((predictions - y_simple) ** 2)
        J_vals[i, j] = cost

# Create 3D surface plot
fig = plt.figure(figsize=(14, 6))

# 3D surface
ax1 = fig.add_subplot(121, projection='3d')
theta0_grid, theta1_grid = np.meshgrid(theta0_vals, theta1_vals)
ax1.plot_surface(theta0_grid, theta1_grid, J_vals.T, cmap='viridis', alpha=0.8)
ax1.scatter([model_normal.theta[0]], [model_normal.theta[1]], 
           [model_gd.cost_history[-1]], color='red', s=100, marker='*', label='Optimum')
ax1.set_xlabel('θ₀ (intercept)', fontsize=10)
ax1.set_ylabel('θ₁ (slope)', fontsize=10)
ax1.set_zlabel('Cost J(θ)', fontsize=10)
ax1.set_title('Cost Function Surface', fontsize=12)
ax1.legend()

# Contour plot
ax2 = fig.add_subplot(122)
contour = ax2.contour(theta0_grid, theta1_grid, J_vals.T, levels=30, cmap='viridis')
ax2.plot(model_normal.theta[0], model_normal.theta[1], 'r*', markersize=15, label='Optimum')
ax2.set_xlabel('θ₀ (intercept)', fontsize=12)
ax2.set_ylabel('θ₁ (slope)', fontsize=12)
ax2.set_title('Cost Function Contours', fontsize=14)
ax2.legend()
plt.colorbar(contour, ax=ax2, label='Cost')

plt.tight_layout()
plt.show()

print("The red star shows the optimal parameters found by both methods!")
```

## 6. Comparing Methods

Let's compare gradient descent and normal equation on different dataset sizes.

```python
import time

# Test different dataset sizes
sizes = [100, 500, 1000, 5000, 10000]
gd_times = []
normal_times = []

print("Comparing methods on different dataset sizes...\n")
print(f"{'Size':<10} {'GD Time (s)':<15} {'Normal Time (s)':<20} {'Faster Method':<15}")
print("="*70)

for size in sizes:
    # Generate data
    X_test = 2 * np.random.rand(size, 1)
    y_test = 3 + 2 * X_test + np.random.randn(size, 1)
    
    # Time gradient descent
    start = time.time()
    model_gd_test = LinearRegressionGD(learning_rate=0.1, n_iterations=100)
    model_gd_test.fit(X_test, y_test)
    gd_time = time.time() - start
    gd_times.append(gd_time)
    
    # Time normal equation
    start = time.time()
    model_normal_test = LinearRegressionNormal()
    model_normal_test.fit(X_test, y_test)
    normal_time = time.time() - start
    normal_times.append(normal_time)
    
    faster = "Gradient Descent" if gd_time < normal_time else "Normal Equation"
    print(f"{size:<10} {gd_time:<15.6f} {normal_time:<20.6f} {faster:<15}")

print("\n✓ Comparison complete")
```

```python
# Plot timing comparison
plt.figure(figsize=(10, 6))
plt.plot(sizes, gd_times, 'o-', linewidth=2, markersize=8, label='Gradient Descent (100 iter)')
plt.plot(sizes, normal_times, 's-', linewidth=2, markersize=8, label='Normal Equation')
plt.xlabel('Dataset Size (m)', fontsize=12)
plt.ylabel('Time (seconds)', fontsize=12)
plt.title('Training Time: Gradient Descent vs Normal Equation', fontsize=14)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.show()

print("\nKey Observations:")
print("- Gradient Descent: Linear time complexity O(m×n×iterations)")
print("- Normal Equation: Cubic time complexity O(n³) due to matrix inversion")
print("- For small datasets: Normal equation is often faster")
print("- For large datasets: Gradient descent scales better")
```

## 7. Key Takeaways

### ✅ What We Learned

1. **Linear Regression Model**:
   - Models linear relationships: $\hat{y} = \theta^T X$
   - Parameters: $\theta_0$ (intercept), $\theta_1, ..., \theta_n$ (slopes)

2. **Cost Function (MSE)**:
   - Measures prediction error: $J(\theta) = \frac{1}{2m} \sum (\hat{y}^{(i)} - y^{(i)})^2$
   - Convex function (single global minimum)

3. **Gradient Descent**:
   - Iterative optimization algorithm
   - Update rule: $\theta := \theta - \alpha \frac{\partial J}{\partial \theta}$
   - Requires choosing learning rate $\alpha$ and iterations
   - Scales well to large datasets

4. **Normal Equation**:
   - Closed-form solution: $\theta = (X^T X)^{-1} X^T y$
   - No iterations needed (one-step solution)
   - Slow for large feature sets ($O(n^3)$)

5. **When to Use What**:
   - **Small datasets (n < 10,000)**: Normal equation
   - **Large datasets (n > 10,000)**: Gradient descent
   - **Very large datasets**: Stochastic/Mini-batch gradient descent
