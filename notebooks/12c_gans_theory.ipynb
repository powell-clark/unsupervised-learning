# Lesson 12C: Generative Adversarial Networks (GANs) - Theory

Learn adversarial training for generation.

**GAN** consists of two networks in competition:
- **Generator G**: Creates fake samples from noise z
- **Discriminator D**: Distinguishes real vs fake

**Training**: Minimax game  
min_G max_D V(D,G) = E[log D(x)] + E[log(1 - D(G(z)))]

**Process:**
1. Train D to maximize correct classification (real=1, fake=0)
2. Train G to fool D (make D(G(z)) → 1)
3. Iterate until equilibrium

## GAN Algorithm

**For each iteration:**

1. **Train Discriminator:**
   - Sample real data x ~ p_data
   - Sample noise z ~ p_z
   - Generate fake data x_fake = G(z)
   - Update D to maximize log D(x) + log(1 - D(x_fake))

2. **Train Generator:**
   - Sample noise z ~ p_z
   - Update G to maximize log D(G(z))

**Convergence:** G learns to match p_data, D outputs 0.5 everywhere.

```
import torch
import torch.nn as nn

class Generator(nn.Module):
    def __init__(self, latent_dim=10, output_dim=2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.ReLU(),
            nn.Linear(32, 32),
            nn.ReLU(),
            nn.Linear(32, output_dim)
        )
    
    def forward(self, z):
        return self.net(z)

class Discriminator(nn.Module):
    def __init__(self, input_dim=2):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 32),
            nn.LeakyReLU(0.2),
            nn.Linear(32, 32),
            nn.LeakyReLU(0.2),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )
    
    def forward(self, x):
        return self.net(x)

# Initialize
G = Generator(latent_dim=10, output_dim=2)
D = Discriminator(input_dim=2)

# Generate from noise
z = torch.randn(5, 10)
fake_samples = G(z)
predictions = D(fake_samples)

print(f"Generated samples shape: {fake_samples.shape}")
print(f"Discriminator predictions: {predictions.squeeze().detach().numpy()}")
print("✅ GAN architecture defined!")
```
