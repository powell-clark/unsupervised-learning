# Lesson 12C: GANs - Theory

**GAN** = Generator vs Discriminator in minimax game.

**Generator**: Creates fake samples from noise
**Discriminator**: Distinguishes real from fake

```
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

torch.manual_seed(42)
print('✅ Loaded!')
```

## Minimax Objective

min_G max_D V(D,G) = E[log D(x)] + E[log(1 - D(G(z)))]

```
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(10, 32), nn.ReLU(), nn.Linear(32, 2))
    
    def forward(self, z):
        return self.net(z)

class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(nn.Linear(2, 32), nn.LeakyReLU(0.2), nn.Linear(32, 1), nn.Sigmoid())
    
    def forward(self, x):
        return self.net(x)

G = Generator()
D = Discriminator()
print('✅ GAN defined')
```

```
# Test forward pass
z = torch.randn(5, 10)
fake = G(z)
pred = D(fake)
print(f'Generated shape: {fake.shape}')
print(f'Discriminator output: {pred.squeeze().detach().numpy()}')
print('✅ GAN architecture working')
```
