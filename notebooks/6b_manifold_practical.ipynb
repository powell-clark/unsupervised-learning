## PCA vs t-SNE vs UMAP Comparison

Apply all three dimensionality reduction methods and compare results.

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import umap

np.random.seed(42)
print("✅ Loaded!")
```

```
# Load data
digits = load_digits()
X, y = digits.data, digits.target

# Apply all three methods
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

reducer = umap.UMAP(n_components=2, random_state=42)
X_umap = reducer.fit_transform(X)

# Compare
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

methods = [
    ('PCA', X_pca, f'Variance: {pca.explained_variance_ratio_.sum():.2%}'),
    ('t-SNE', X_tsne, 'Perplexity: 30'),
    ('UMAP', X_umap, 'n_neighbors: 15')
]

for ax, (name, X_reduced, subtitle) in zip(axes, methods):
    scatter = ax.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='tab10', 
                        s=30, alpha=0.7, edgecolors='k')
    ax.set_title(f'{name}\n{subtitle}', fontweight='bold', fontsize=12)
    ax.set_xlabel('Component 1', fontweight='bold')
    ax.set_ylabel('Component 2', fontweight='bold')
    ax.grid(True, alpha=0.3)

plt.colorbar(scatter, ax=axes, label='Digit')
plt.suptitle('Dimensionality Reduction Comparison: Digits Dataset (64D → 2D)',
            fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()

print("✅ PCA: Linear, fast, interpretable")
print("✅ t-SNE: Nonlinear, preserves local structure, slow")
print("✅ UMAP: Nonlinear, fast, preserves local+global structure")
```
