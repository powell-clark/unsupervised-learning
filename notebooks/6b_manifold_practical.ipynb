# Lesson 6B: Manifold Learning - Practical\n\nCompare t-SNE and UMAP for visualization.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
np.random.seed(42)
plt.style.use('seaborn-v0_8-darkgrid')
print("✅ Loaded!")
```

```
from sklearn.manifold import TSNE
import umap
from sklearn.datasets import load_digits

digits = load_digits()
X, y = digits.data, digits.target

# t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# UMAP
reducer = umap.UMAP(random_state=42)
X_umap = reducer.fit_transform(X)

# Compare
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=20)
axes[0].set_title('t-SNE')
axes[1].scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='tab10', s=20)
axes[1].set_title('UMAP')
plt.show()
print("✅ Both methods visualize high-D data in 2D!")
```
