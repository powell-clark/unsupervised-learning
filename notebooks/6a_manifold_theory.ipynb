# Lesson 6A: Manifold Learning - Theory

## The High-Dimensional Visualization Challenge

You're analyzing **single-cell RNA sequencing data** from 10,000 cells, each with 20,000 gene expression measurements. You need to understand cell types and developmental trajectories, but:
- Can't visualize 20,000 dimensions
- PCA loses too much nonlinear structure
- Need to preserve local neighborhoods

**Manifold hypothesis**: High-dimensional data often lies on a low-dimensional manifold.

**Solution**: Manifold learning methods like **t-SNE** and **UMAP** that preserve local structure while reducing to 2D/3D for visualization.

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
import umap

np.random.seed(42)
plt.style.use('seaborn-v0_8-darkgrid')
print("✅ Loaded!")
```

## t-SNE: t-Distributed Stochastic Neighbor Embedding

**Key idea**: Preserve pairwise similarities in low dimensions

**Algorithm**:
1. Compute pairwise similarities in high-D using Gaussian kernel
2. Compute pairwise similarities in low-D using t-distribution
3. Minimize KL divergence between the two distributions

**Perplexity**: Balances local vs global structure (typical: 5-50)

```
# Compare t-SNE perplexity
digits = load_digits()
X, y = digits.data, digits.target

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for idx, perplexity in enumerate([5, 30, 50]):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_tsne = tsne.fit_transform(X[:500])
    
    ax = axes[idx]
    scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y[:500], cmap='tab10', s=30, alpha=0.7)
    ax.set_title(f't-SNE (perplexity={perplexity})', fontweight='bold')
    ax.grid(True, alpha=0.3)

plt.colorbar(scatter, ax=axes, label='Digit')
plt.suptitle('t-SNE: Effect of Perplexity', fontweight='bold', fontsize=14)
plt.tight_layout()
plt.show()
print("✅ t-SNE perplexity comparison complete!")
```

## UMAP: Uniform Manifold Approximation and Projection

**Advantages over t-SNE**:
- Faster (can handle millions of points)
- Better preserves global structure
- Can transform new points

**Key parameters**:
- n_neighbors: Local neighborhood size (default: 15)
- min_dist: Minimum distance in embedding (default: 0.1)

```
# UMAP demonstration
reducer = umap.UMAP(n_components=2, n_neighbors=15, random_state=42)
X_umap = reducer.fit_transform(X)

plt.figure(figsize=(12, 10))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1], c=y, cmap='tab10', s=30, alpha=0.7, edgecolors='k')
plt.colorbar(scatter, label='Digit Class')
plt.title('UMAP: Digits Dataset (64D → 2D)', fontweight='bold', fontsize=14)
plt.xlabel('UMAP 1', fontweight='bold')
plt.ylabel('UMAP 2', fontweight='bold')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
print("✅ UMAP visualization complete!")
```
