# Lesson 6A: Manifold Learning (t-SNE, UMAP) - Theory\n\n**Manifold learning** reduces dimensions while preserving local structure.

**t-SNE**: t-distributed Stochastic Neighbor Embedding preserves local distances.
**UMAP**: Uniform Manifold Approximation preserves both local and global structure.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
np.random.seed(42)
plt.style.use('seaborn-v0_8-darkgrid')
print("✅ Loaded!")
```

## t-SNE Algorithm

1. Compute pairwise similarities in high dimensions
2. Initialize low-dimensional embedding
3. Optimize to match similarities using gradient descent

Better than PCA for non-linear data!

```
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

digits = load_digits()
X, y = digits.data, digits.target

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='tab10', s=20)
plt.colorbar(scatter)
plt.title('t-SNE on Digits Dataset', fontweight='bold')
plt.show()
print("✅ t-SNE reduced 64D to 2D!")
```
